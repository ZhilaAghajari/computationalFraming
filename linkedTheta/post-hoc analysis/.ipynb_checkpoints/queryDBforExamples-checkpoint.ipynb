## Insert each documents' words, with their reln and argument2 into DB

```python
query = "INSERT INTO parsed_documents(document_id,word,relns,arg2) where word is NULL"\
"VALUES (%s, %s, %s, %s)"
# args =(df['document_id'],['word'], df['relns'], df['arg2'])
for index, row in df.iterrows():
    if (type(row['word']) == float and pd.isna(row['word'])) or row['relns'] is None or row['arg2'] is None:
        continue
    args = (row['document_id'], row['word'], row['relns'], row['arg2'])

    cursor.execute(query, args)

cursor.execute(query, args)
sql_db.commit()
# cursor.close()
# sql_db.close()
```

```python
def docs4q(word,reln,arg,cursor):
    '''
    for a (word, reln, arg) find a document in which thee co-occur
    '''
    query = "select * from parsed_documents where word='{0}' AND relns='{1}' AND arg2='{2}'".format(word,reln,arg)
    cursor.execute(query)
    res = cursor.fetchall()
    return pd.DataFrame(res,columns=['document_id','word','relns','arg2'])

#test with one example
docs4q('bob','nn.dep','wheaton',cursor)
```

```python
#Read the topic document probablity resulted from theta role model. this file includes "topic_id", "document_id" and "probablity_score"
topic_doc_df = pd.read_csv('../topicDocProbabilities_5_topic_5_thetarole_sample_15_prepCollapsed.csv')
topic_doc_df.columns, len(topic_doc_df)
```

## Post-hoc analysis: get topic terms, their relns, args, and document id in which they occur.

```python
# Post-hoc analysis: get topic terms, their relns, args, and document id in which they occur.
import numpy as np
from tqdm import tqdm
topic_ids = df_topics.topic_id.unique().tolist()
# topic_ids = [2, 3, 4]
ls = []
for tid in tqdm(topic_ids):
    print('*** topic id: {0} ***'.format(tid))
    print('* top terms {0}'.format(' ,'.join((df_topics[df_topics.topic_id==tid].word).tolist())))
    words = df_topics.loc[df_topics.topic_id==tid].word.tolist()
    for w in words:
        #get reln and args
        relns = df_topics.loc[(df_topics.topic_id==tid) & (df_topics.word==w)].relns.str.split(',').tolist()[0]
        arg2 = df_topics.loc[(df_topics.topic_id==tid) & (df_topics.word==w)].arg2.str.split(',').tolist()[0]
#         print('* w : ', w)
#         print(' *** Realns {1}\n'.format(w, relns))
#         print(' *** Args {1}\n'.format(w, arg2))
        #get document ids for these words' in reln and with arg2....
        for reln in relns:
            for arg in arg2:
                #this is the only place that can change to read from DB insetad of dataframe..
                #tdf = df[(df.word == w) & (df.relns.str.contains(reln)) & (df.arg2.str.contains(arg))]
                tdf= docs4q(w,reln,arg,cursor)
                tls = []
                cand_ids = [idd for idd,rl,rg in zip(list(tdf.document_id),list(tdf.relns),list(tdf.arg2)) \
                           if np.sum([1 for r,a in zip(rl.split(','),rg.split(',')) if r == reln and a == arg])>=1.0]
                cand_ids = topic_doc_df[(topic_doc_df.document_ids.isin(cand_ids)) & \
                                       (topic_doc_df.topic_id == tid) & \
                                        (topic_doc_df.probability_score>=0.7)].document_ids.tolist() # depenting on the porptoion of 
                if len(cand_ids)>0:
                    print('topic -- {0}, word -- {1}, reln -- {2}, arg -- {3}, docs: {4}'.format(tid,w,reln,arg,cand_ids))
                    ls.append({'topic_id':tid,'word':w,\
                  'relns':reln,\
                  'arg2':arg, 'document_ids':cand_ids})
                

res = pd.DataFrame(ls)
res.to_csv('post_hoc_examples_5_topics_5_thetarole.csv')
```
